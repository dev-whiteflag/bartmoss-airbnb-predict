```python
import os
import cv2
```

```python
vcap = cv2.VideoCapture('siege-iq.mp4')
success, image = vcap.read()

count = 0
while success:
    vcap.set(cv2.CAP_PROP_POS_MSEC, (count * 50))
    success,image = vcap.read()
    
    if success is True:
        cv2.imwrite('./img/frame%d.jpg' % count, image)
        print('saved frame num ', count)
        count = count + 1
```

```python
for root, dirs, files in os.walk(os.path.abspath("./img")):
    count = 0;
    for file in files:
        image = cv2.imread(os.path.join(root, file))
        crop = image[290:1390, 1017:1560]
        cv2.imwrite('./img/crop/frame%d-cut.jpg' % count, crop)
        print('saved frame num cropped ', count)
        count = count + 1
```
